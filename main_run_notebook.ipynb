```python
import networkx as nx
from text_filterer import formTokens
from n_gram_graph import gram3Generator
from clustering_coefficient import wordsByClusteringCoefficient
from betweenness_centrality import wordsByBetweennessCentrality
from degree_centrality import wordsByDegreeCentrality
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
```

```python
G=gram3Generator('FilteredTexts/Port.txt')
```

```python
wordsByClusteringCoefficient(G)
```

```python
wordsByDegreeCentrality(G)
```

```python
#################################### Betweenness Centrality ##########################################
BC_dict=dict(nx.betweenness_centrality(G))

nodesizearray=[]
for n in G.nodes:
	nodesizearray.append(10000*BC_dict[n])

figure(num=None, figsize=(50, 50), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(121)
nx.draw(G,node_size=nodesizearray, with_labels=True, font_weight='bold')
######################################################################################################
```

```python
###################################   Degree Centrality Plot  ########################################
de_dict=dict(nx.degree_centrality(G))

nodesizearray=[]
for n in G.nodes:
	nodesizearray.append(10000*de_dict[n])

figure(num=None, figsize=(50, 50), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(121)
nx.draw(G,node_size=nodesizearray, with_labels=True, font_weight='bold')
######################################################################################################
```
